```python
#---------------- visualization
sns.set_palette('colorblind')
```

```python
#------- Salary distribution-----------------
plt.figure(figsize=(10,5))
sns.histplot(hrdf['salary'], bins=30, kde=True)
plt.title("Distribwution of Employee's Salaries")
plt.xlabel("Salary")
plt.ylabel("Count")
plt.show()
```

```python
#------- job satisfaction distribution-----------------
plt.figure(figsize=(10,5))
sns.countplot(data=hrdf, x="jobsatisfaction", hue="jobsatisfaction", legend=False)
plt.title("job satisfaction of Employees")
plt.xlabel("job satisfaction Level")
plt.ylabel("No. of Employees")
plt.show()
```

```python
#------- Attriion-----------------
plt.figure(figsize=(10,5))
sns.countplot(data=hrdf, x="attrition", hue="attrition", legend=False)
plt.title("attrition count in the company")
plt.xlabel("attrition (yes/no)")
plt.ylabel("count")
plt.show()
```

```python
#------- work-----------------
plt.figure(figsize=(10,5))
sns.lineplot(data=hrdf, x="worklifebalance", y="jobsatisfaction", color='darkblue')
plt.title("work lifebalance Vs job satisfaction")
plt.xlabel("work life balance")
plt.ylabel("job satisfaction")
plt.show()
```

```python
#What is distribution of employees by gender
empByGender=hrdf['gender'].value_counts()
empByGender
```

```python
plt.figure(figsize=(12,6))
plt.pie(empByGender,
        labels=empByGender.index,
        autopct="%.2f%%", 
        textprops={'size':'large', 
                   'fontweight':'bold',
       'rotation':20, 'color':'w'})
plt.title('distribution of employees by gender', fontsize=18, fontweight='bold')
plt.legend(bbox_to_anchor=(1, 1))
plt.show()
```

```python
#2.	What is Salary distribution by job role
salarybyRole=hrdf.groupby('jobrole')['salary'].median()
salarybyRole
```

```python
# Create boxplot
plt.figure(figsize=(12, 8))
ax = sns.boxplot(x='jobrole', y='salary', data=hrdf)
plt.title('Salary Distribution by Job Role', fontsize=16)
plt.xlabel('Job Role', fontsize=12)
plt.ylabel('Salary ($)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
